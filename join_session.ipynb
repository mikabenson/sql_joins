## USING clause 

```python
# Here we can use the clause to join the tables only if the column name for both tables are identical
# For this case the column_name (productCode) is identical and appears in both tables
# To join orderdetails and products on productCode with the using() clause
query1 ="""
SELECT *
FROM orderdetails
     JOIN products
     USING (productCode)
     LIMIT 10;
"""
pd.read_sql(query1, conn)
```

# SQL JOINS..

```python
join_L = """
SELECT *
FROM products
     LEFT JOIN orderdetails
     USING (productCode);
"""
df4 = pd.read_sql(join_L, conn)
df4
```

```python
print("Number of records returned:", len(df4))
print("Number of records where order details are null:", len(df4[df4.orderNumber.isnull()]))
```

```python
# To join customers using the alias c with employees using the alias e on the 
# foreign keys salesRepEmployeeNumber and employeeNumber and order the result by employeeNumber
# then type the code to execute the query:
query = """
SELECT *
  FROM customers AS c
       JOIN employees AS e
       ON c.salesRepEmployeeNumber = e.employeeNumber
       ORDER By employeeNumber;
"""
pd.read_sql(query, conn)
```

```python
E = """
SELECT *
   FROM customers c
   JOIN employees e
   ON c.salesRepEmployeeNumber = e.EmployeeNumber
   ORDER BY employeeNumber;
"""
pd.read_sql(E, conn)
```

```python
#  Here we join customers using the alias c with employees using the alias e on 
# the foreign keys salesRepEmployeeNumber and employeeNumber and order the result by employeeNumber, 
# then type the code to execute the query:
sales_query = """
SELECT * 
  FROM customers c
       JOIN employees AS e
       ON c.salesRepEmployeeNumber = e.employeeNumber
       ORDER BY employeeNumber;
"""
df5 = pd.read_sql(sales_query, conn)
df5
```

```python
print("Number of records returned:", len(df5))
print("Number of records where employees have no customer number are null:", len(df5[df5.customerNumber.isnull()]))
```
